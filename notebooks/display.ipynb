```python
import numpy as np
from scipy.signal import find_peaks, find_peaks_cwt

def find_period_autocorrelation(signal):
    # Удаляем DC смещение
    signal_centered = signal - np.mean(signal)
    
    # Вычисляем автокорреляцию
    autocorr = np.correlate(signal_centered, signal_centered, mode='full')
    autocorr = autocorr[len(autocorr)//2:]  # Берем только правую часть
    
    # Находим пики (кроме первого)
    peaks, _ = find_peaks(autocorr)
    #print(len(peaks))

    return peaks
    if len(peaks) < 2:
        raise ValueError("Недостаточно пиков для определения периода")
    print(len(peaks))
    # Первый пик после нуля соответствует периоду
    return peaks[1] - peaks[0]
```

```python
from sklearn.metrics import classification_report
from numba import njit 
@njit()
def r2_score(y_true, y_pred):
    
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    if y_true.shape != y_pred.shape:
        raise ValueError(f"Формы y_true и y_pred должны совпадать. {y_true.shape}, {y_pred.shape}")
    
    y_mean = np.mean(y_true)
    ss_total = np.sum((y_true - y_mean)**2)
    ss_residual = np.sum((y_true - y_pred)**2)
    
    if ss_total == 0:
        # Если все истинные значения одинаковы
        return 1.0 if ss_residual == 0 else 0.0
    return 1 - (ss_residual / ss_total)
    
@njit()
def mean_squared_error(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    if y_true.shape != y_pred.shape:
        raise ValueError("Формы y_true и y_pred должны совпадать.")
    return np.mean((y_true - y_pred)**2)

@njit()
def rmse(y_predict, y_real):
        return np.sqrt(mean_squared_error(y_predict, y_real))


    
scores = {"R²":r2_score, "RMSE":rmse, "Report": classification_report}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import HistGradientBoostingClassifier

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV


dataset = datasets[0]
dataset.split_data(test_size=0.2, valid_size=0.01)
model = HistGradientBoostingClassifier(random_state=42)
```

```python
def create_time_windows(X, y, window_size=3):
    X_new, y_new = [], []
    for i in range(len(X) - window_size):
        X_window = X.iloc[i:i+window_size].values.flatten()  # "Разворачиваем" окно в одну строку
        y_label = y.iloc[i+window_size]  # Берём метку для последнего элемента в окне
        X_new.append(X_window)
        y_new.append(y_label)
    return np.array(X_new), np.array(y_new)
train_x, train_y = create_time_windows(dataset.X_train, dataset.Y_train, 10)
test_x, test_y = create_time_windows(dataset.X_test, dataset.Y_test, 10)
```

```python

model.fit(train_x, train_y)
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_iter': [100, 200, 300],
    'max_depth': [3, 5, 7, None],
    'min_samples_leaf': [1, 5, 10],
    'l2_regularization': [0.0, 0.1, 0.5],
    'max_bins': [128, 255],
    'early_stopping': [True, False]
}
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_grid,
    n_iter=50,  # количество случайных комбинаций
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1,
    verbose=4,
    random_state=42
)

# Запуск поиска
random_search.fit(train_x, train_y)

best_model = random_search.best_estimator_

for score in scores:
    print(f"{score} = {scores[score](best_model.predict(test_x), test_y)}")
```
