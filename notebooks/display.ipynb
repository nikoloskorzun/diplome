```python
import numpy as np
from scipy.signal import find_peaks, find_peaks_cwt

def find_period_autocorrelation(signal):
    # Удаляем DC смещение
    signal_centered = signal - np.mean(signal)
    
    # Вычисляем автокорреляцию
    autocorr = np.correlate(signal_centered, signal_centered, mode='full')
    autocorr = autocorr[len(autocorr)//2:]  # Берем только правую часть
    
    # Находим пики (кроме первого)
    peaks, _ = find_peaks(autocorr)
    #print(len(peaks))

    return peaks
    if len(peaks) < 2:
        raise ValueError("Недостаточно пиков для определения периода")
    print(len(peaks))
    # Первый пик после нуля соответствует периоду
    return peaks[1] - peaks[0]
```

```python
from sklearn.metrics import classification_report
from numba import njit 
@njit()
def r2_score(y_true, y_pred):
    
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    if y_true.shape != y_pred.shape:
        raise ValueError(f"Формы y_true и y_pred должны совпадать. {y_true.shape}, {y_pred.shape}")
    
    y_mean = np.mean(y_true)
    ss_total = np.sum((y_true - y_mean)**2)
    ss_residual = np.sum((y_true - y_pred)**2)
    
    if ss_total == 0:
        # Если все истинные значения одинаковы
        return 1.0 if ss_residual == 0 else 0.0
    return 1 - (ss_residual / ss_total)
    
@njit()
def mean_squared_error(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    if y_true.shape != y_pred.shape:
        raise ValueError("Формы y_true и y_pred должны совпадать.")
    return np.mean((y_true - y_pred)**2)

@njit()
def rmse(y_predict, y_real):
        return np.sqrt(mean_squared_error(y_predict, y_real))


    
scores = {"R²":r2_score, "RMSE":rmse, "Report": classification_report}
```
